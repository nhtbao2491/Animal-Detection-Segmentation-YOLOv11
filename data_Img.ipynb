<h4 style="color: orange">2. Thống kê tập dữ liệu </h4>

**📌 Dataset Animals 151 có những gì ?**


````python
import os

dataset = "./Dataset"

def count_images(folder_path):
    total_images = 0
    for _, _, files in os.walk(folder_path):
        if len(files) == 1:
            continue
        num_files = len(files)
        total_images += num_files
    return total_images
slImg = count_images(dataset)
print(f"Tổng số lượng ảnh trong tập dữ liệu: {slImg}")
````

````python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("./Dataset/train/Lion/Lion_11.jpg")
h , w, c = img.shape
print(f"Kích thước ảnh: {h}x{w} pixels")
print("Số kênh màu (channels): ", c)

# Hiển thị ảnh bằng matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # chuyển đổi màu từ BGR -> RGB
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
````

````python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("./Dataset/train/Snow_Leopard/Snow_Leopard_15.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(image)

fig, axes = plt.subplots(1, 4, figsize=(15, 5))
axes[0].imshow(image)
axes[0].set_title("Ảnh gốc")
axes[1].imshow(R, cmap="Reds")
axes[1].set_title("Kênh Red")
axes[2].imshow(G, cmap="Greens")
axes[2].set_title("Kênh Green")
axes[3].imshow(B, cmap="Blues")
axes[3].set_title("Kênh Blue")

for ax in axes:
    ax.axis("off")

plt.show()
````
