## I. Chia dữ liệu thành train & val

```python
import os
import random
import shutil

source_dir = "./Create_newLabel/Output_Folder/"
train_dir = "./Dataset/train"
val_dir = "./Dataset/val"

for label in os.listdir(source_dir):
    label_path = os.path.join(source_dir, label)
    if not os.path.isdir(label_path):
        continue
    images = [f for f in os.listdir(label_path) if os.path.isfile(os.path.join(label_path, f))]
    random.shuffle(images)
    val_images = images[:2]
    train_images = images[2:]
    train_label_path = os.path.join(train_dir, label)
    val_label_path = os.path.join(val_dir, label)
    os.makedirs(train_label_path, exist_ok=True)
    os.makedirs(val_label_path, exist_ok=True)
    for img in val_images:
        src = os.path.join(label_path, img)
        dst = os.path.join(val_label_path, img)
        shutil.copy(src, dst)
    for img in train_images:
        src = os.path.join(label_path, img)
        dst = os.path.join(train_label_path, img)
        shutil.copy(src, dst)
print("Đã hoàn tất chia dữ liệu vào train/ và val/ theo yêu cầu.")
```

## II. Tạo Danh Sách Label

```python
import os
source_dir = "./Dataset/train"
output_file = "DS_Labels.txt"
labels = sorted([label for label in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, label))])
with open(output_file, "w", encoding="utf-8") as f:
    for idx, label in enumerate(labels):
        f.write(f"{label}\n")
print(f"Đã ghi danh sách label vào {output_file}")
```

## III. Biến đổi ảnh

```python
from PIL import Image
import os

input_folder = './Transform_Img/Input_Img'
output_folder = './Transform_Img/Output_Img'

def transform_Image(input_folder, output_folder):
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.png', '.jpeg', '.jpg', '.bmp', '.webp')):
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path).convert('RGB')
            img_resized = img.resize((640, 640))
            base_name = os.path.splitext(filename)[0]
            output_path = os.path.join(output_folder, base_name + '.jpg')
            img_resized.save(output_path, format='JPEG', quality=95)
            os.remove(img_path)
    print("Hoàn tất chuyển đổi và resize ảnh.")
    
transform_Image(input_folder, output_folder)
```

## IV. Tạo 1 label mới

```python
import os
import shutil

def create_newLabel(label_name):
    input_dir = './Create_newLabel/Input_Img'
    output_dir = './Create_newLabel/Output_Folder'
    label_folder = os.path.join(output_dir, label_name)
    if not os.path.exists(label_folder):
        os.makedirs(label_folder)
    image_files = [f for f in os.listdir(input_dir) if f.lower().endswith(('jpg', 'jpeg', 'png'))]
    for idx, image_file in enumerate(image_files, 1):
        image_path = os.path.join(input_dir, image_file)
        new_name = f"{label_name}_{idx}.jpg"
        new_image_path = os.path.join(label_folder, new_name)

        shutil.move(image_path, new_image_path)
        print(f"Đã di chuyển và đổi tên {image_file} thành {new_name}")
    
create_newLabel('Rhino')
```

## V. Copy ảnh

```python
import os
import shutil

def copy_images_only(source_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for root, _, files in os.walk(source_dir):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp')):
                src_path = os.path.join(root, file)
                dst_path = os.path.join(output_dir, file)
                shutil.copy2(src_path, dst_path)
    print(f"Đã copy toàn bộ ảnh sang '{output_dir}'.")

source_dir = './Dataset/val'          
output_dir = './Input_Train/images/val'        
copy_images_only(source_dir, output_dir)
```

## VII.Đánh giá mô hình

```python
from ultralytics import YOLO
import numpy as np

best_model_Detection = './runs/detect/train/weights/best.pt'
best_model_Segmentation = './runs/segment/train/weights/best.pt'

detect = YOLO(best_model_Detection)
segment = YOLO(best_model_Segmentation)

metrics_detect = detect.val(data='./Input_Train/Animals_BoundingBox-2/data.yaml', imgsz=640, batch=16, save=False)
metrics_segment = segment.val(data='./Input_Train/Animals_Polygon-1/data.yaml', imgsz=640, batch=16, save=False)

```

```python
def print_metrics(metrics):
    results_dict = metrics.results_dict
    precision = results_dict['metrics/precision(B)']
    recall = results_dict['metrics/recall(B)']
    f1_score = 2 * (precision * recall) / (precision + recall + 1e-6)
    mAP50 = results_dict['metrics/mAP50(B)']
    mAP5095 = results_dict['metrics/mAP50-95(B)']
    iou = mAP5095

    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1_score:.4f}")
    print(f"mAP@0.5: {mAP50:.4f}")
    print(f"mAP@0.5:0.95: {mAP5095:.4f}")
    print(f"IoU (theo mAP50-95): {iou:.4f}")
print("----------------------- Detection -----------------------")
print_metrics(metrics_detect)
print("----------------------- Segmentation -----------------------")
print_metrics(metrics_segment)
```
